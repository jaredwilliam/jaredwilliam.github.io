```python
data = dict(type = 'choropleth',
           locations = df['iso_code'],
           z = df['daily_vaccinations_per_million'],
           text = df['country'],
           colorbar = {'title':'Daily Vaccinations per Million'})
```

```python
layout = dict(title = 'Coronavirus Daily Vaccinations per Million',
             geo = dict(showframe=False, projection = {'type':'mercator'}))
```

```python
choromap3 = go.Figure(data=[data], layout=layout)
choromap3.update_layout(height=800, margin={"r":0,"t":0,"l":0,"b":0})
```
